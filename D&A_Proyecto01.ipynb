<h2>Ejercicio 1:</h2>
Escribir un programa que pregunte al usuario por las ventas de los
últimos 5 días, y muestre por pantalla una serie con los datos de
las ventas por producto: Nro de unidades y precio unitario,
indexada por los días, antes y después de aplicarles un descuento
del 5%.

Desarrollo 1:

```python
producto1 = input('Ingrese el nombre del primer producto: ' )

unidades = []
for i in range(1, 6):
    unidades.append(float(input('Introduce las unidades vendidas en el día ' + str(i) +': ')))

ventas = []
for i in range(1, 6):
    ventas.append(float(input('Introduce el monto de venta del día ' + str(i) +': ')))

print('Serie con las unidades vendidas de los últimos 5 días:')
serie_unidades = pd.Series(unidades, index=['día 1', 'día 2', 'día 3', 'día 4', 'día 5'])
print(serie_unidades)

print('Serie con las ventas de los últimos 5 días antes de aplicar el descuento del 5%:')
serie_ventas1 = pd.Series(ventas, index=['día 1', 'día 2', 'día 3', 'día 4', 'día 5'])
print(serie_ventas1)

print('Serie con las ventas de los últimos 5 días después  de aplicar el descuento del 5%:')
serie_ventas2 = pd.Series(ventas, index=['día 1', 'día 2', 'día 3', 'día 4', 'día 5'])
print(serie_ventas2*0.95)
```

<h2>Ejercicio 2</h2>
Escribir un programa que reciba un diccionario con los ingresos
mensuales de los empleados de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.

```python
cantidad_empleados = int(input('¿Cuántos empleados va a registrar?: '))
nombre = {}
ingresos = {}
for i in range(cantidad_empleados):
  nombre[i] = input('ID del empleado nro ' + str(i+1) + ':')
  ingresos[i] = input('Ingresos del empleado nro ' + str(i+1) + ':')

nombre_valores = nombre.values()
nombre_lista = list(nombre_valores)

ingresos_valores = ingresos.values()
ingresos_lista = list(ingresos_valores)

serie_ingresos = pd.Series(ingresos_lista, index=[nombre_lista])
print(serie_ingresos)
print('El ingreso mínimo es: ', min(serie_ingresos))
print('El ingreso máximo es: ', max(serie_ingresos))
print('La media de ingresos es: ', serie_ingresos.median())
```
